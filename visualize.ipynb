```python
from matplotlib import pyplot as p 
import numpy as np
```

```python
x=np.array([10,20])
y=np.array([20,40])
```

```python
p.plot(x,y)
```

```python
p.plot(x,y,"o")
```

```python
a=np.array([10,30,50,60])
b=np.array([100,200,300,400])
```

```python
p.plot(a,b,"o")
```

```python
p.plot(b)
```

```python
p.plot(x,y,marker="+")
```

```python
p.plot(x,y,marker="*")
```

```python
p.plot(x,y,'p:k')
```

```python
p.plot(x,y,'p-k')
```

```python
p.plot(a,b,"*:m")
```

```python
p.plot(x,y,'p--r')
```

```python
p.plot(x,y,'p--r',ms=10)
```

```python
p.plot(a,b,'*--r',ms=50)
```

```python
p.plot(x,y,'p--r',ms=50,mfc='k')
```

```python
p.plot(x,y,'1--r',ms=65)
```
